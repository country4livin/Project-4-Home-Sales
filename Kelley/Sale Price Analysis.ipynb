```python
#Import Dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv("kingcountysales_2020_2023 (1).csv")
```

```python
# Drop null values
df = df.dropna(subset=["sale_price", "city", "sqft"])
df = df[(df["sale_price"] > 0) & (df["sqft"] > 0)]
```

```python
# Box Plot for Sale Price by Top 15 Cities
# Set up the figure and axes
fig, ax = plt.subplots(figsize=(12, 6))
top_cities = df["city"].value_counts().nlargest(15).index
sns.boxplot(ax=ax, data=df[df["city"].isin(top_cities)], x="city", y="sale_price")
ax.set_title("Sale Price Distribution by City (Top 15 Cities)")
ax.set_ylabel("Sale Price")
ax.set_xlabel("City")
ax.tick_params(axis='x', rotation=45)
# Show Box Plot
plt.tight_layout()
plt.show()
plt.savefig("Output/sale_price_distribution_by_city.png")
```

```python
# Scatter Plot with Regression Line for Sale Price vs. Square Footage
plt.figure(figsize=(10, 6))
sns.regplot(
    data=df,
    x="sqft",
    y="sale_price",
    scatter_kws={"alpha": 0.5},
    line_kws={"color": "red"}
)
plt.title("Sale Price vs. Square Footage with Regression Line")
plt.xlabel("Square Footage")
plt.ylabel("Sale Price")
plt.tight_layout()
plt.show()
plt.savefig("Output/sale_price_vs_sqft.png")
```
