```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Read in the CSV file
file = Path('/Users/fern/Desktop/Bootcamp_Stuff/Project_4/kingcountysales_2000_2023.csv')
df = pd.read_csv(file)
```

```python
# Preview the data
df.head()
```

```python
# Review the data types
print(df.dtypes)
```

```python
# Change the data type of the sale_date column from an object to date time
df['sale_date'] = pd.to_datetime(df['sale_date'])
print(df.dtypes)
```

```python
df.head()
```

```python
# Drop the data from 1999 to 2019 
df = df[~df['sale_date'].dt.year.between(1999, 2019)]
df.head()
```

```python
# drop the 'Unnamed column'
df = df.drop(columns=['Unnamed: 0', 'pinx', 'sale_warning', 'wfnt', 'golf', 'greenbelt', 'noise_traffic', 'view_rainier', 'view_olympics', 'view_cascades',
             'view_territorial', 'view_skyline', 'view_sound', 'view_lakewash', 'view_lakesamm', 'view_otherwater', 'view_other', 'submarket'])
df.set_index('sale_id')
df.head()
```

```python
df.to_csv('./kingcountysales_2020_2023.csv', index=False)
```
