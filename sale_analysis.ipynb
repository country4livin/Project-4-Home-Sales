```python
#Import Dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv("Resources/st_paul_sold_properties.csv")
```

```python
# Compute Pearson correlation coefficient after dropping NaNs
clean_df = df[["sqft", "sold_price"]].dropna()
correlation_value, p_value = pearsonr(clean_df["sqft"], clean_df["sold_price"])
print(f"Pearson Correlation Coefficient: {correlation_value:.2f} (p-value: {p_value:.4f})")
```

```python
# Scatter Plot with Regression Line for Sale Price vs. Square Footage
plt.figure(figsize=(10, 6))
sns.regplot(
    data=df,
    x="sqft",
    y="sold_price",
    scatter_kws={"alpha": 0.5},
    line_kws={"color": "red"}
)
plt.title("Sale Price vs. Square Footage with Regression Line")
plt.xlabel("Square Footage")
plt.ylabel("Sold Price")
plt.tight_layout()
plt.savefig("output/sale_price_vs_sqft.png")
plt.show()
```
